```python
model_id = "sentence-transformers/all-MiniLM-L6-v2"
# "get your token in http://hf.co/settings/tokens"
hf_token =''
```

```python
import pandas as pd
df = pd.read_csv('dataset_pu_dg_token_terbaru.csv')
```

```python
df
```

```python
data = df['tokens'].astype(str)
```

```python
data = pd.DataFrame({"tokens" : data})
```

```python
kateg = df['kategori']
```

```python
pasal = df['pasal']
```

```python
kontent = df['konten']
```

```python
tok = data['tokens']
```

```python
katgri = kateg.to_list()
```

```python
psl = pasal.to_list()
```

```python
kntn = kontent.to_list()
```

```python
tokn= tok.to_list()
```

```python
import requests

api_url = f"https://api-inference.huggingface.co/pipeline/feature-extraction/{model_id}"
headers = {"Authorization": f"Bearer {hf_token}"}
```

```python
def query(texts):
    response = requests.post(api_url, headers=headers, json={"inputs": texts, "options":{"wait_for_model":True}})
    return response.json()
```

```python
output_kategori= query(katgri)
```

```python
output_pasal=query(psl)
```

```python
output_konten= query(kntn)
```

```python
output_tokens = query(tokn)
```

```python
import numpy as np
import pandas as pd
```

```python
pasal = pd.DataFrame(psl)
kategori = pd.DataFrame(katgri)
```

```python
combination3 = np.column_stack([output_kategori, output_pasal, output_konten, output_tokens])
```

```python
emn = pd.DataFrame(combination3)
```

```python
emn
```

```python
emn['kategori'] = kategori[0]
```

```python
emn['pasal'] = pasal[0]
```

```python
col_0 = emn.pop('kategori')
emn.insert(0, 'kategori', col_0)
```

```python
col_1 = emn.pop('pasal')
emn.insert(1, 'pasal', col_1)
```

```python
emn
```

```python
emn.to_csv('doc_pu_with_embeddings.csv', index=False)
```
